## 2. Discover a real truth yourself

The safest, most satisfying early wins are **things you find in the data** — un-leakable, and they *are* the core lesson. Run the three cells below. Each is ~10 lines of pandas and each overturns a common SEO belief.

Every number is **computed live from the shipped CSV** — nothing is hardcoded.

```python
import pandas as pd, numpy as np
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(df.shape[0], "rows,", df.shape[1], "columns")
df.head(3)
```

### Discovery A — "High search volume means more traffic." Does it?

```python
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")
print("Near zero -> keyword search volume barely predicts the traffic a page actually gets.")
```

### Discovery B — the CTR cliff by position
Click-through rate is not flat: it collapses as you move down the results.

```python
visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print(ctr_by_pos.round(4).to_string())
ctr_by_pos.plot(kind="bar", title="Mean CTR by position tier (impressions >= 100)", ylabel="CTR");
```

### Discovery C — is longer content the lever?
Compare word count for **declining** vs **growing** pages.

```python
wc = df.groupby("trend_direction")["word_count"].median()
print(wc.round(0).to_string())
print("\n'down' vs 'up' pages have almost the same median word count -> length is not the lever.")
```

### My Discovery — Are search volume and impressions correlated?
Originally in Discovery A, the dataset included pages with 0 impressions.

What does 0 impression means?

It means that Google never showed that page in search results during the last 90 days which might have happened due to many reasons such as lower page ranking, poor optimisation, high competition etc.

These pages may affect the correlation, so I wanted to see whether removing them changes the relationship between search volume and impressions.


```python
filtered = df[df["impressions_90d"] > 0]

corr = filtered["search_volume"].corr(filtered["impressions_90d"])

print(f"Correlation between search_volume and impressions_90d (impressions > 0): {corr:.3f}")
print("Still near zero -> keyword search volume barely predicts the traffic a page actually gets.")
```

# My Observation
I repeated the Discovery A after filtering while the dataset only keeping pages with impressions_90d > 0. But even after that the correlation between search_volume and impressions_90d remained almost unchanged (0.001). This suggests that in this dataset removing pages with zero impressions does not strengthen the relationship between search volume and page impressions. Since search volume alone appears to be a weak predictor of impressions, other factors are likely influencing page visibility. This is an observed pattern in this sample and should not be interpreted as a general rule.
