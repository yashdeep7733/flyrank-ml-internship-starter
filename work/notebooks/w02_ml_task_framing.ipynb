# Lane: Refresh / Content Opportunity Scoring

I frame this as a scoring task. The goal is to assign each webpage a score representing how likely it is to benefit from a content refresh. Pages with higher scores can be prioritized by the content team.

Although this can also be viewed as a ranking problem, scoring is the most natural framing because every page receives a continuous value that can be used to rank refresh opportunities.

```python
TASK_TYPE = "Scoring"

print("ML Task:", TASK_TYPE)
```

## 2. Target or proxy

# Target / Proxy

The goal is to predict a content refresh score for each webpage. Since the dataset does not contain a direct label indicating whether a page needs refreshing, a proxy based on search performance trends can be used. Features such as trend_direction, trend_pct, impressions_90d, impressions_last_30d, clicks_90d, and clicks_last_30d provide signals about whether a page’s performance is improving or declining. Pages with stronger signs of declining performance would receive higher refresh scores.

```python
import pandas as pd

df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

print("Possible target-related columns:")

target_columns = [
    "trend_direction",
    "trend_pct",
    "impressions_90d",
    "impressions_last_30d",
    "clicks_90d",
    "clicks_last_30d"
]

df[target_columns].head()
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule, such as refreshing every page with a negative trend percentage, is too simplistic because content performance depends on many factors. Search volume, competition, historical impressions, clicks, content characteristics, and performance trends all influence whether refreshing a page is worthwhile. A machine learning approach can learn patterns from multiple features simultaneously and assign a refresh score based on their combined effect. This helps prioritize pages more effectively than a single rule or threshold.

```python
# Features that could influence refresh priority

features = [
    "search_volume",
    "competition",
    "word_count",
    "impressions_90d",
    "clicks_90d",
    "trend_pct"
]

df[features].head()
```
