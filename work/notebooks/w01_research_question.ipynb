# Lane - Refresh / Content Opportunity Scoring

I chose this lane because it addresses a practical problem faced by content teams: deciding which webpages should be reviewed first when time and resources are limited. For example, a  website may have 30,000 pages but only 3 writers. How will they decide which pages should be given the attention first? Instead of manually inspecting thousands of pages one by one, We can take the help of machine learning which will analyze multiple observable search and engagement signals to help reviewers know which pages should be given attention first.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

print(f"Rows: {len(df)}")
print(f"Columns: {len(df.columns)}")

corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.4f}")
```

The starter dataset contains **30,000 webpages** with **44 features**, providing enough information to explore patterns related to search performance. I also observed that the correlation between `search_volume` and `impressions_90d` is **0.0012**, which suggests that search volume alone does not explain page impressions. This indicates that multiple signals should be considered when prioritizing webpages for manual review, making the Refresh / Content Opportunity Scoring lane suitable for machine learning.
